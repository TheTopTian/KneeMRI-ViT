## Using the Dino v2 to add the mask

```python
import numpy as np
import matplotlib.pyplot as plt

dataset = np.load("mask.npy")
out = np.load("../MRNet/MRNet-v1.0/train/sagittal/0000.npy")
```

```python
def plot_slices(array, color="gray", title=None):
	n_slices = array.shape[0]
	n_rows = int(n_slices / 4)
	fig, axes = plt.subplots(nrows=n_rows, ncols=4, figsize=(20, 20))
	plt.axis('off')
	for i, ax in enumerate(axes.flatten()):
		if i < n_slices:
			ax.imshow(array[i, :, :], cmap=color) # viridis
			ax.set_title(f'Slice {i}')
			ax.axis('off')
	if title:
		fig.suptitle(title, fontsize=16)
	plt.show()
```

```python
def save_slices(array,color="gray"):
    for i in range(array.shape[0]):
        fig,ax = plt.subplots()
        plt.axis('off')
        ax.imshow(array[i],cmap="gray")
        fig.savefig(f'../separate_slices/{i}.png', transparent = True, bbox_inches = 'tight', pad_inches = 0)
        plt.close(fig)
```

```python
plot_slices(dataset)
```

```python
save_slices(out)
```
