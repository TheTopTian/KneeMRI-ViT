# Check how many cases with location errors

```python
import pandas as pd
import ast

location_csv = pd.read_csv('../../new_location.csv',header=0)
pathologys = location_csv.columns.tolist()

def read_csv(view,disease_num,pathologys):
    if view == "tranversal":
        depth = 27
    else:
        depth = 25
    df = pd.read_csv(f"./location_diseases/disease{disease_num}_location.csv")
    datas = df[view].tolist()
    print(f"There are {len(datas)} cases with location in disease: {pathologys[disease_num]}")
    datas = [ast.literal_eval(x) for x in datas]

    x_value = []
    y_value = []
    z_value = []
    for data in datas:
        x_value.append(data[0])
        y_value.append(data[1])
        z_value.append(data[2])
    
    count = 0
    for i in z_value:
        if i >= depth:
            count += 1
    
    print(f"{count} cases have the location error")
```

```python
views = ["coronal","sagittal","transversal"]
disease_num = 1
while disease_num <14:
    for view in views:
        read_csv(view,disease_num,pathologys)
    disease_num += 1
```
